```python
#Neural Net Regression Approach - Scikit's optimized CART algirithm
#Mitchell Rhead 10185289
#April 2022
#sachinrajput, “sklearn Neural Network,” Kaggle.com, 04-Sep-2020. [Online]. Available: https://kaggle.com/sachinrajput17/sklearn-neural-network. [Accessed: 1-Apr-2022].

import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import seaborn as sns

dataset = pd.read_excel('regression_data.xlsx')
x =dataset.iloc[:,0:9]
y =dataset.iloc[:,9]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33, random_state=42)

rr = Ridge(alpha=0.01)
rr.fit(x_train, y_train) 
pred_train_rr= rr.predict(x_train)
print('training mse is %.8f' % mean_squared_error(y_train,pred_train_rr))
print('training variance(r^2) score is %.8f' % r2_score(y_train, pred_train_rr))

pred_test_rr= rr.predict(x_test)
print('test mse is %.8f' % mean_squared_error(y_test,pred_test_rr)) 
print('test variance(r^2) score is %.8f' % r2_score(y_test, pred_test_rr))

#scatter
plt.scatter(y_test,pred_test_rr,color="black",marker='.')
sns.regplot(x=y_test, y=pred_test_rr,marker='.')
plt.title("Ridge Regression")
plt.ylabel('Predicted')
plt.show()
```
